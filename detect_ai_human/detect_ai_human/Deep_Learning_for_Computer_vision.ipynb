```python

import os
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image

preprocess_input = tf.keras.applications.mobilenet_v3.preprocess_input

# Paths
MODEL_PATH = "C:/Users/Christian/Desktop/AI_or_human/best_model_mobilenetv3.keras"
TEST_CSV_PATH = "C:/Users/Christian/Desktop/AI_or_human/AI_or_human/data/test.csv"
TEST_IMAGES_PATH = "C:/Users/Christian/Desktop/test_data_v2"
OUTPUT_CSV_PATH = "C:/Users/Christian/Desktop/AI_or_human/submission.csv"

# Load the model
model = tf.keras.models.load_model(MODEL_PATH)

# Load the test CSV
test_df = pd.read_csv(TEST_CSV_PATH)

# Image preprocessing function for testing
def preprocess_image(img_path, target_size=(512, 512)):
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = preprocess_input(img_array)  # Use MobileNetV3 preprocessing
    return img_array

# Make predictions
predictions = []
for img_path in test_df['id']:
    img_filename = os.path.basename(img_path)  # Extract only the filename
    full_img_path = os.path.join(TEST_IMAGES_PATH, img_filename)

    if not os.path.exists(full_img_path):  # Debugging check
        print(f"File not found: {full_img_path}")

    img_array = preprocess_image(full_img_path)
    pred = model.predict(img_array)[0][0]  # Get prediction score
    label = 1 if pred > 0.5 else 0
    predictions.append(label)

# Save predictions to CSV
submission_df = pd.DataFrame({'id': test_df['id'], 'label': predictions})
```
